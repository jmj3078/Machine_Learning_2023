```python
train = pd.read_csv('mnist_train.csv', header=None)
X_train = np.array(train.iloc[:, 1:].values.tolist())
y_train = np.array(train[0].values.tolist())
```

```python
test = pd.read_csv('mnist_test.csv', header=None)
X_test = np.array(test.iloc[:, 1:].values.tolist())
y_test = np.array(test[0].values.tolist())
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

## MLP from Scratch

- 설정된 learning rate, batch size, neuron, epoch, layer 수 등은 변경하지 말 것!

```python
from sklearn import metrics
from my_mlp import *
import time

mlp = my_MLP()

start = time.time()
mlp.fit(X_train, y_train)

taken = time.time() - start
print(taken, '\n')

y_pred = mlp.predict(X_test)

print(metrics.classification_report(y_test, y_pred))
```
